## Cluster Analysis

```
#import module
from sklearn.cluster import KMeans

# Filter data
bp_emp =  dtnew[['LastAchievemet', 'Achievement3quartal']][dtnew.bestp == 1]

# Create groups using K-means clustering.
kmeans = KMeans(n_clusters = 3, random_state = 0).fit(bp_emp)
```

```
# Add new column "label" annd assign cluster labels.
bp_emp['label'] = kmeans.labels_
# Draw scatter plot
plt.scatter(bp_emp['LastAchievemet'], bp_emp['Achievement3quartal'], c=bp_emp['label'],cmap='Accent')
plt.xlabel('Penghargan Terakhir')
plt.ylabel('Poin Penghargaan')
plt.title('Penghargaan yang diraih')
plt.show()
```
